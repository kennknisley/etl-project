```python
import os
import requests
import pandas as pd
import time
import json
import csv
```

```python
base_url='https://financialmodelingprep.com/api/v3/financials/income-statement/'
dow_tickers=['AAPL', 'AXP', 'BA', 'CAT', 'CSCO', 'CVX', 'DIS', 'DOW', 'GS', 'HD', 'IBM',
             'INTC', 'JNJ', 'JPM', 'KO', 'MCD', 'MMM', 'MRK', 'MSFT', 'NKE', 'PFE', 'PG',
             'TRV', 'UNH', 'V', 'VZ', 'WBA', 'WMT', 'XOM']
```

```python
Tick_Symbol=[]
Date=[]
Revenue=[]
Revenue_Growth=[]
Cost_of_Revenue=[]
Gross_Profit=[]
RD_Expenses=[]
SGA_Expenses=[]
Operating_Expenses=[]
Operating_Income=[]
Interest_Expense=[]
Earning_before_Tax=[]
Income_Tax_Expense=[]
Earnings_before_Tax=[]
Income_Tax_Expense=[]
Net_Income=[]
EPS_Diluted=[]
Weighted_Average_Shs_Out=[]
Dividend_Per_Share=[]
Gross_Margin=[]
Profit_Margin=[]
Consolidated_Income=[]
Net_Profit_Margin=[]
```

```python
for comp in dow_tickers:
    try:
        query_url=f'{base_url}{comp}'
        dow_response=requests.get(query_url).json()
        for x in range(0,3):
            try:
                Tick_Symbol.append(dow_response['symbol'])
                Date.append(dow_response['financials'][x]['date'])
                Revenue.append(dow_response['financials'][x]['Revenue'])
                Revenue_Growth.append(dow_response['financials'][x]['Revenue Growth'])
                Cost_of_Revenue.append(dow_response['financials'][x]['Cost of Revenue'])
                Gross_Profit.append(dow_response['financials'][x]['Gross Profit'])
                RD_Expenses.append(dow_response['financials'][x]['R&D Expenses'])
                SGA_Expenses.append(dow_response['financials'][x]['SG&A Expense'])
                Operating_Expenses.append(dow_response['financials'][x]['Operating Expenses'])
                Operating_Income.append(dow_response['financials'][x]['Operating Income'])
                Interest_Expense.append(dow_response['financials'][x]['Interest Expense'])
                Earnings_before_Tax.append(dow_response['financials'][x]['Earnings before Tax'])
                Income_Tax_Expense.append(dow_response['financials'][x]['Income Tax Expense'])
                Net_Income.append(dow_response['financials'][x]['Net Income'])
                EPS_Diluted.append(dow_response['financials'][x]['EPS Diluted'])
                Weighted_Average_Shs_Out.append(dow_response['financials'][x]['Weighted Average Shs Out'])
                Dividend_Per_Share.append(dow_response['financials'][x]['Dividend per Share'])
                Gross_Margin.append(dow_response['financials'][x]['Gross Margin'])
                Consolidated_Income.append(dow_response['financials'][x]['Consolidated Income'])
                Net_Profit_Margin.append(dow_response['financials'][x]['Net Profit Margin'])
            except AttributeError as e:
                print(e)

        time.sleep(1)
    except AttributeError as e:
        print(e)
        
        
```

```python
dow_df=pd.DataFrame({'Ticker Symbol':Tick_Symbol, 'Date':Date, 'Revenue':Revenue, 'Revenue Growth': Revenue_Growth,
                    'Cost of Revenue': Cost_of_Revenue, 'Gross Profit': Gross_Profit, 'R&D Expenses': RD_Expenses,
                    'SG&A Expenses': SGA_Expenses, 'Operating Expenses': Operating_Expenses, 'Operating Income': Operating_Income,
                    'Interest Expense': Interest_Expense, 'Earnings before Tax': Earnings_before_Tax, 'Income Tax Expense': Income_Tax_Expense,
                    'Net Income': Net_Income, 'EPS Diluted': EPS_Diluted, 'Weighted Average Shs Out': Weighted_Average_Shs_Out,
                    'Divident Per Share': Dividend_Per_Share,'Gross Margin': Gross_Margin,'Consolidated Income': Consolidated_Income,
                    'Net Profit Margin': Net_Profit_Margin})
dow_df
```
